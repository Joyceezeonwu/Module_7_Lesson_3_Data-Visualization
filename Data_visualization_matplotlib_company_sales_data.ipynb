```python
#importing necessary libraries

import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import numpy as npy
```

```python
df = pd.read_csv('company_sales_data.csv')
print(df)
```

```python
#Comparing any two products sales data using a bar chart.

x = npy.arange(len(df['month_number']))
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
fig = plt.figure(figsize=(14, 9))
ax = df.plot.bar(x = 'month_number', y=['shampoo', 'moisturizer'], figsize = (14,9),  width = 0.7, color=['red', 'violet'])
plt.legend(loc = 'upper right')
plt.title("JORA Group Product Comparison")
plt.xlabel('Month')
plt.ylabel("Sales of Facewash and Shampoo")
plt.xticks(x, month)
plt.savefig('JORA_Group.png')
plt.show()
```

```python
#Assuming the average monthly sales of bathing soap is 10k, visualize the sales data of bathing soap of all months, showing how they differ from the average sale.
fig = plt.figure(figsize=(14, 6))
x = npy.arange(len(df['month_number']))
month = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
yerr = npy.array(list(map(lambda x: 10000-x, df['bathingsoap'])))
yerr = npy.absolute(yerr)

y = df['bathingsoap']
plt.bar(x,y, yerr=yerr, capsize=10, label = 'bathingsoap', color='green')

plt.title('Sales Data of Bathing Soap')
plt.xlabel('Month')
plt.ylabel("Sales of Toiletries")
plt.xticks(x, month)
plt.legend()
plt.savefig('Sales Data of Bathing Soap.png')
plt.show()
```
